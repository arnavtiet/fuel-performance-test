```python
import pandas as pd
from pycaret.regression import *
from sklearn.model_selection import train_test_split


data = pd.read_csv('./Fuel_cell_performance_data-Full.csv')

X = data.filter(regex='^F')
y = data['Target1']

print(y)
print(X)
```

```python



class_counts = y.value_counts()
low_count_classes = class_counts[class_counts < 2].index


filtered_data = data[~data['Target1'].isin(low_count_classes)]






```

```python
# Split the data into 70% training and 30% testing using stratified sampling
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


train_data = pd.concat([X_train, y_train], axis=1)
test_data = pd.concat([X_test, y_test], axis=1)

```

```python
# Initialize PyCaret regression setup
clf = setup(data=train_data,
            target='Target1',
            session_id=123,
            verbose=False)
```

```python

best_model = compare_models()


```

```python
final_model = finalize_model(best_model)
```

```python
predictions = predict_model(final_model, data=test_data)
```
